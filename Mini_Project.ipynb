```python
import pandas as pd
student_info=pd.read_csv('/content/student_performance_dataset_50.csv')
student_info
```

```python
input_dataset=student_info.drop(columns=['InternetAtHome','FinalResult'])
input_dataset
```

```python
output_dataset=student_info[['InternetAtHome','FinalResult']]
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
student_model=DecisionTreeClassifier()
student_model.fit(input_dataset,output_dataset)
Student_grade_point=student_model.predict([[2.0,5,1000,10,60,12]])
Student_grade_point
```
